```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

## List all cities where the area is greater than 5000, along with their population.

```python
q = ''' select  c.name CityName, c.population CityPopulation  from facts f 
        left join cities c 
        on 
        f.id = c.facts_id
       where Area > 5000
        
    '''

sql_query(q)
```

## List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%

```python
q = ''' select  c.name CityName  from facts f 
        right join cities c 
        on 
        f.id = c.facts_id
       where f.death_rate > 2 and f.birth_rate < 20 and c.capital = 1
       
        
    '''

sql_query(q)
```

## List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth.

```python
q = ''' select  c.name CityName,f.population_growth  from facts f 
        right join cities c 
        on 
        f.id = c.facts_id
       where f.population_growth between 0.01 and 0.05
       order by 2 desc
    '''

sql_query(q)
```

## Find countries that start with &quot;A&quot; and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join

```python
q = '''select f.name CountryName, c.name CityName from facts f
       left join cities c on
       f.id = c.facts_id
       where f.name like 'A%' and c.name is null
       
       
       '''
sql_query(q)
```

## List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
q = '''select c.name CityName, f.death_rate from facts f
       left join cities c on
       f.id = c.facts_id
       where c.capital = 1
       
       '''
sql_query(q)
```
